```python
import jieba.posseg as pseg
from collections import Counter
from collections import defaultdict
import random
```

```python
line_no = 0
limit = 100
lm = Counter()

ngram = 3

for line in open('happiness.txt'):
    line_no += 1
    if line_no > limit:
        break
    words = pseg.cut(line.strip())
    words = [i.word for i in words] ## 转成list，取词不要词性
    
    dct = defaultdict(Counter)
    
    for i in range(ngram-1, len(words)):
        context = words[(i-ngram):i]
        word = words[i]  
        dct['context']['word'] += 1

        print("\t".join(context))
        print(word)
s = float(sum(lm.values()))
for key, cnt in lm.items():
    lm[key] /= s        print("----")
```

```python
## 归一
s = float(sum(lm.values()))
for key, cnt in lm.items():
    lm[key] /= s

    for word in cnt:
        cnt[word] /= s  # works in Python 3
            
```

```python
def gen_w(context):
    """根据 context 及条件概率，随机生成 word"""
    r = random.random()
    psum = 0
    for word, prob in dct[context].items():
        psum += prob
        if psum > r:
            return word
```

```python
def gen_s(m):
    
    """生成 m 个句子"""
    sentences = []
    text = ''
    context = tuple([START for i in range(ngram-1)])
    i = 0
    while (i < m):
        word = gen_w(context)
        text = str(text) + str(word)
        context = tuple((list(context) + [word])[1:])
    ##    if word in BREAK:
    ##        sentences.append(text)
    ##        text = ''
    ##        context = tuple([START for i in range(ngram-1)])
    ##        i += 1
    return sentences

```
